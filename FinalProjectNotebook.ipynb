### Analysis
As seen from the bar plots, the average of Fenty Beauty, Makeup Forever, and U.S. Best Sellers ranges from 60-70 Lightness values. In contrast, the average of Nigeria, Japan, and India's Best Sellers ranges from 50 to 76, with Japan and India ranging from 70-80. These results remain constant with Western and Eastern beauty standards that I previously stated in the Introduction. It is important to note that beauty products produced in Nigeria, Japan, and India are most likely produced for their respective monoethnic countries. This may indicate certain skintone ranges are prioritized in the production of foundation shades. However, skintones in each ethnicity come in large ranges and should be included in makeup. In addition, as beauty products become more globalized, beauty companies may strive to be inclusive of more demographics. This is seen in the Western makeup shade averages as these regions have more populations with various demographics.  

After bar plots, the scatter plots are created using the saturation and lightness of the foundation shades. The scatter plots of Fenty Beauty show that their foundation shade collection is evenly distributed, emphasizing that the brand does not focus on a certain skintone range. The scatter plots of Makeup Forever are mainly clustered within the 65-85 Lightness value range with data gaps on both ends of the range. This indicates the Makeup Forever's foundation range does not cater to individuals that are on either ends of the foundation shade range. The scatter plots of U.S. best sellers fall within the 60-80 value range, a lighter skin-tone range. 

The scatter plots of Nigeria best-sellers are widely dispersed and have less plots compared to other graphs. It is important to note that Nigeria has the least amount of entries, affecting how representative the dataset for this particular category. The scatter plots of Japan's best-sellers are realtively evenly distributed. However, there is a data gap from value 70 and below, highlighting that there are fewer foundation shade options for deeper skin tones. The scatter plot of India's best-sellers are mainly within the 70-80 value range. Below the value 70, there are very few plots. This emphasized that India's best-selling foundation products are mainly catered towards lighter skin tones. 

Overall, it can be seen that the Western brand categories do have darker foundation shade options compared to Japan and India. This finding is in line with the Western beauty standard of tan skin. In contrast, Japan and India's foundation shade range is more limited and clusters around light foundation shades. These findings can reveal the priorities of each respective beauty industries and whether or not they are making an effort to be more inclusive.  

### Big Picture
The discussion of inclusivity is important in foundation shade range. Because of how limited many brands' foundation shade range is, many people of color are left excluded and have difficulty finding a foundation match as skin colors and skin tones are more complex than what is available. The Western best-sellers and Nigeria have lower averages, indicating that a large portion of their foundation shade entries consits of darker foundation shades compared to Japan and India. Japan and India's best-sellers are more limited in shade range and prioritize light/lighter skin tones. This finding is consistent with the Asian beauty standards of preferring lighter skin tones. And while Japan and India may have less diversity, these best-selling brands could still produce foundation shades for over-sea consumers interested in trying their products. 


### Findings Useful?
These visualizations and analyses emphasize how people of color are still excluded in many foundation shade ranges regardless of location as many brands have data clusters around lighter skin tones. The conversation of inclusivity in makeup was on the rise after Rihanna created Fenty Beauty and released the PRO Filt'r Foundation with 40 shades. Afterwards, many makeup-users have decided whether or not to support brands that do not put effort towards including consumers of darker skin colors. In addition, skin tones, such as olive, yellow, pink, etc. have also started to be widely considered when expanding foundation shade range. While this is dataset is from 2018, it is still relatively recent. It emphasizes that a majority of makeup products still lack a wide foundation shade range. This downfall may also indicate how the brands' other products such as eyeshadow, blush, lipstick, etc. also may not be pigmented enough for darker skin colors as it can be seen that inclusivity is not an important of concern. Ultimately, it points out how even though producing a large foundation shade range is definitely feasible as seen from Fenty, many brands choose to leave out many shades, showing lack of consumer empathy.

### Analysis
As seen from the bar plots of BIPOC recommended products from BIPOC founded brands versus BIPOC recommended from White founded brands. The average of White founded brands is 66.401786 while the average of the BIPOC recommended brands is 
56.035294. This emphasizes that BIPOC recommended brands have a larger shade range, including more darker shades ranges compared to White founded brands. In the scatter plots, BIPOC founded plots are relatively evenly distributed, implying that these brands have put in equal amount of effort in producing shades in all skin color ranges. In contrast, White founded plots are mainly concentrated around Lightness values 60-80, or light/lighter foundation shades. These results points out that a brand's founders play a role in how inclusive foundation shade ranges are, which may come from personal experiences with skin color/skin tone that BIPOC have. 

### Big Picture
As seen from the data visualization and average of lightness, BIPOC founded brands have a more extensive and evenly distributed foundation shade range compared to White founded brands. BIPOC owners may have personal experience with struggling to find the correct foundation match. In contrast, White owners do not experience the extent of this as many brands mainly cater towards creating light/lighter shade foundations as seen from the statistics and graphs above. These findings emphasize how beauty products made by individuals that come from similar backgrounds and demographics understand the significance of having a wide foundation shade range. 


### Findings Useful?
These findings emphasize how a beauty brand's founders do make a difference in the extent of inclusivity of their shade range. This points out how diversity within a company plays a large role as it provides more perspectives and voices that are representative of the general public. With a diverse company including darker skin shades, POC would not have to inconvienence themselves when trying to find the correct shade match. By having more POC exectutives on the board of beauty companies, they can point out flaws and inconsistencies in the products that non-POC may not see or have an issue with. In addition, a more diverse board would be representative of the demographics of real consumers. Entrepreneurship within the beauty industry should have inclusitivity be a standard when creating products, not an afterthought. The inclusion of more foundation shades would not only speak volumes on what the brand stands for, but also attract a larger consumer base as consumers would want to support companies that also support them. The gradual expansion of foundation shades in recent times has allowed for more diversity in the beauty community, but this increase in inclusitivity also needs to be seen in other products such as eyeshadow, lipsticks, contour etc. POC makeup users often struggle to have these products appear pigmented and vivid on their skin. This sentiment can also be extended to the models and influencers that represent the brand and their products. Consumers of all demographics would be able to understand and imagine how the product would look on their skin color/ skin tone based off advertisements and product demos. Overall, it is extremely important to empathize with the consumer in every step of product development for customers to feel considered and cared for.
